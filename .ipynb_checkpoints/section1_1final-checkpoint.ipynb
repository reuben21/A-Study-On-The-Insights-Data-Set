```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('Insights.csv')
df.columns
```

```python
df.dropna(inplace = True)
df.isna().sum()
```

```python
df_filtered = df.copy()
```

```python
df_most_performing = df_filtered.groupby('Product').sum()
```

```python
df_most_performing
```

```python
df_most_performing['Profit'] = df_most_performing['Value(INR)']-(df_most_performing['Qty']*df_most_performing['Value(USD)'])
```

```python
df_most_performing
```

```python
df_most_performing.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
langs = df_most_performing['Product']
students = df_most_performing['Profit']
ax.bar(langs,students)
plt.show()
```
