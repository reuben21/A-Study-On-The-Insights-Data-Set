```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('Insights.csv')
df.columns
```

```python
df.dropna(inplace = True)
```

```python
df.isna().sum()
```

```python
len(df)
```

```python
from geopy.geocoders import ArcGIS
```

```python
gc=ArcGIS(scheme='https')
```

```python
df.head(100)
```

```python
df_small = pd.DataFrame()
```

```python
df_small
```

```python
for i in range(0,4800,100):
    df_temp = df[i:i+100].copy()
    df_temp["coordinates"]=df_temp["Port of Destination"].apply(gc.geocode)
    df_small = df_small.append(df_temp)
    print(i)
df_temp = df[4800:4971].copy()
df_temp["coordinates"]=df_temp["Port of Destination"].apply(gc.geocode)
df_small = df_small.append(df_temp)
df_small
```

```python
df_small['Latitude'] = df_small['coordinates'].apply(lambda x: x.latitude if x != None else None)
df_small['Longitude'] = df_small['coordinates'].apply(lambda x: x.longitude if x != None else None)
```

```python
df_small
```

```python
df_small.to_csv('location.csv') 
```
