```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('Insights.csv')
df.columns
```

```python
df.dropna(inplace = True)
df.isna().sum()
```

```python
df_filtered = df.copy()
```

```python
df_most_performing = df_filtered.groupby('Product').sum()
```

```python
df_most_performing
```

```python
df_most_performing['Profit'] = df_most_performing['Value(INR)']-(df_most_performing['Qty']*df_most_performing['Value(USD)'])
```

```python
df_most_performing
```

```python
df_most_performing.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df_most_performing.index
```

```python
prod = df_most_performing.index.tolist()
profit = df_most_performing['Profit'].tolist()
```

```python
prod_new = prod[0:10]
profit_new = profit[0:10]
d1 = {}
for l1_ in prod_new:
    for l2_ in profit_new:
        d1[l1_] = l2_
        profit_new.remove(l2_)
        break
dictionary = {k: v for k, v in sorted(d1.items(), key=lambda item: item[1])}
dictionary
```

```python
print(profit)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(dictionary.keys(),dictionary.values())
plt.show()
```
